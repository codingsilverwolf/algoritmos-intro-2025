```python
import pandas as pd
import random

nombres_comunes = ['Juan', 'María', 'Pedro', 'Laura', 'Carlos', 'Sofía', 'Martín', 'Valentina', 'Lucas', 'Agustina',
                   'Mateo', 'Camila', 'Nicolás', 'Florencia', 'Santiago', 'Julieta', 'Sebastián', 'Lucía', 'Tomás', 'Martina',
                   'Benjamín', 'Emilia', 'Joaquín', 'Antonella', 'Gabriel', 'Pilar', 'Felipe', 'Abril', 'Bruno', 'Olivia',
                   'Thiago', 'Catalina', 'Máximo', 'Victoria', 'Lautaro']
apellidos_comunes = ['González', 'Rodríguez', 'López', 'Pérez', 'Martínez', 'Gómez', 'Sánchez', 'Díaz', 'Romero', 'Suárez',
                     'Fernández', 'Torres', 'Vargas', 'Benítez', 'Flores', 'Acosta', 'Ruiz', 'Castro', 'Silva', 'Herrera',
                     'Moreno', 'Núñez', 'Bianchi', 'Ferrari', 'Russo', 'Espósito', 'Ricci', 'Colombo', 'Vega', 'Pereyra',
                     'Mendoza', 'García', 'Gutierrez', 'Juárez', 'Luna']

n_alumnos = 35

# Generar listas separadas de nombres y apellidos aleatorios
nombres = [random.choice(nombres_comunes) for _ in range(n_alumnos)]
apellidos = [random.choice(apellidos_comunes) for _ in range(n_alumnos)]

# Generar DNI aleatorios
dnis = random.sample(range(35000000, 40000001), n_alumnos)

n_min_nota = 2  # Define el valor mínimo para las notas
primer_parcial = [random.randint(n_min_nota, 10) for _ in range(n_alumnos)]
segundo_parcial = [random.randint(n_min_nota, 10) for _ in range(n_alumnos)]

# Crear el DataFrame con DNI como índice y columnas separadas para Nombre y Apellido
df_alumnos = pd.DataFrame({'Nombre': nombres,
                           'Apellido': apellidos,
                           'primer_parcial': primer_parcial,
                           'segundo_parcial': segundo_parcial},
                          index=dnis)

# Crear las columnas vacías restantes
df_alumnos['promoción'] = None
df_alumnos['recuperatorio'] = None
df_alumnos['final'] = None
df_alumnos['condicion'] = None

df_alumnos
```

### Si queremos aplicar una operación por filas, lo mejor es utilizar el método .apply con la opción axis = 1 

### df.apply(mi_funcion_sobre_filas, axis = 1) pasará a mi_funcion_sobre_filas cada fila del df.

```python
# Ejemplo 1: supongamos que queremos ver el mínimo de la nota

def min_nota(fila):
    return pd.Series({'Min Parcial': min(fila['primer_parcial'], fila['segundo_parcial'])})

df_alumnos.apply(min_nota, axis = 1)

```

```python
def min_nota_con_nombre_separado(fila):
    min_parcial = min(fila['primer_parcial'], fila['segundo_parcial'])
    apellido = fila['Apellido']
    nombre = fila['Nombre']
    return pd.Series({'Apellido': apellido, 'Nombre': nombre, 'Min Parcial': min_parcial})

df_alumnos.apply(min_nota_con_nombre_separado, axis=1)

```

```python
import pandas as pd

data = {'Nombre': ['Ana', 'Juan'],
        'Edad': [25, 30],
        'Ciudad': ['Buenos Aires', 'Córdoba']}
df = pd.DataFrame(data)

def mostrar_fila(row):
    print(f"Fila con índice {row.name}:")
    print(f"  Nombre: {row['Nombre']}")
    print(f"  Edad: {row['Edad']}")
    print(f"  Ciudad: {row['Ciudad']}")
    print("-" * 20)
    # No es necesario devolver nada si solo quieres visualizar
    # pero .apply() espera un valor de retorno (puede ser None)
    return None

df.apply(mostrar_fila, axis=1)
```

```python
def separar_apellido_nombre(fila):
    """
    Separa el apellido y el nombre de la columna 'Alumno'.
    """
    alumno = fila['Alumno']
    partes = alumno.split(', ')
    apellido = partes[0]
    nombre = partes[1]
    return pd.Series({'Apellido': apellido, 'Nombre': nombre})
    

# Aplicar la función para crear las nuevas columnas
nuevas_columnas = df.apply(separar_apellido_nombre, axis=1)

# Concatenar las nuevas columnas al DataFrame original
df[['Apellido', 'Nombre']] = nuevas_columnas

# Eliminar la columna 'Alumno' original
df.drop('Alumno', axis=1, inplace=True)

df
```

```python
# Crear un DataFrame con las nuevas columnas y valores iniciales pd.NA
nuevas_notas = pd.DataFrame({
    'primer_parcial': pd.NA,
    'segundo_parcial': pd.NA,
    'final': pd.NA,
    'recuperatorio': pd.NA,
    'condicion': pd.NA
}, index=df.index)  # Aseguramos que el índice coincida

# Asignar este DataFrame a las nuevas columnas en df
df[['primer_parcial', 'segundo_parcial', 'final', 'recuperatorio', 'condicion']] = nuevas_notas

df
```
